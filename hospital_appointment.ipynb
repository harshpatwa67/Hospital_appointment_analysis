```python

appointment['appointment_date'] = pd.to_datetime(appointment['appointment_date'])                                                  
treatment['treatment_date'] = pd.to_datetime(treatment['treatment_date'])


billing['bill_date'] = pd.to_datetime(billing['bill_date'])


patient['date_of_birth'] = pd.to_datetime(patient['date_of_birth'])
patient['registration_date'] = pd.to_datetime(patient['registration_date'])
```

```python
appointment.dtypes
```

```python
appointment['appointment_month'] = appointment['appointment_date'].dt.to_period('M')
```

```python
appointment[['appointment_date', 'appointment_month']].head()
```

```python
from datetime import datetime

patient['age'] = (
    (pd.Timestamp('today') - patient['date_of_birth'])
    .dt.days // 365
)
```

```python
patient[['date_of_birth', 'age']].head()
```

```python
patient['age_group'] = pd.cut(
    patient['age'],
    bins=[0, 18, 30, 45, 60, 120],
    labels=['0-18', '19-30', '31-45', '46-60', '60+']
)
```

```python
patient[['age', 'age_group']].head()
```

```python
appointment['appointment_hour'] = (
    pd.to_datetime(
        appointment['appointment_time'],
        format='%H:%M:%S'
    ).dt.hour
)

```

```python
appointment[['appointment_time', 'appointment_hour']].head()
```

```python
for col in appointment.columns:
    if str(appointment[col].dtype).startswith("period"):
        appointment[col] = appointment[col].astype(str)
```
